#### In this notebook, we will use Matplotlib, a foundational data visualization library in Python. The objective is to create clear and informative plots to visually represent key insights from the Video Game Sales dataset. We will generate a variety of charts, including a bar chart to compare publishers, a pie chart to show the market share of different regions, and a histogram to understand the distribution of game releases over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the dataset
df = pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')
df.dropna(inplace=True) # Drop missing values for cleaner plots
```

# --- Matplotlib Visualizations ---

# 1. Bar Chart: Top 10 Publishers by Global Sales

```python
top_publishers = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 7))
plt.bar(top_publishers.index, top_publishers.values, color='skyblue')
plt.title('Top 10 Publishers by Global Sales')
plt.xlabel('Publisher')
plt.ylabel('Total Global Sales (in Millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 2. Pie Chart: Sales Distribution by Region
# Calculate total sales for each major region

```python
region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(8, 8))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title('Global Sales Distribution by Region')
plt.ylabel('') # Hides the 'None' label on the y-axis
plt.show()
```

# 3. Histogram: Distribution of Game Release Years

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Year'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Video Game Release Years')
plt.xlabel('Year')
plt.ylabel('Number of Games Released')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
